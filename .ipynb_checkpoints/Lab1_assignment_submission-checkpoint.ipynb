```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt 
```

# 1. Take a head shot picture of yourself 

## I. Read and display the image 

```python
im_cv = cv2.imread("Arun1.jpg") #Read the image
#cv reads image as bgr by default
im = cv2.cvtColor(im_cv, cv2.COLOR_BGR2RGB)#convert from bgr to rgb and save
plt.axis('off')  # Hide axis
plt.imshow(im) 
```

## II. Print the coordinates.

```python
height, width = im.shape[:2]
channels = im_cv.shape[2] if len(im_cv.shape) == 3 else 1

# Print the width and height
print(f"Image Width: {width} pixels")
print(f"Image Height: {height} pixels")
print(f"Number of Channels: {channels}")
```

## III. Print the picture intensity

```python
# Get pixel intensity values in BGR at the selected point
pixel1_values_bgr = im_cv[300, 300]  # Note: OpenCV uses (y, x) for array indexing
pixel2_values_bgr = im_cv[600, 400]  # Note: OpenCV uses (y, x) for array indexing
# Convert BGR to RGB format
pixel1_values_rgb = pixel1_values_bgr[::-1]
pixel2_values_rgb = pixel2_values_bgr[::-1]
 # Create a string with the pixel values
pixel1_values_str = f"RGB values at 300,300: {pixel1_values_rgb}"
print(pixel1_values_str)
pixel2_values_str = f"RGB values at 400,600: {pixel2_values_rgb}"
print(pixel2_values_str)
```

## Select a point on image with mouse click and know the RGB values of that pixel

### Note:  You have to Run the code to see the output. New window will open after you run the code

```python
# Callback function for mouse events
def on_mouse_event(event, x, y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        # Create a copy of the original image
        im_with_text = im_cv.copy()

        # Get pixel intensity values in BGR at the selected point
        pixel_values_bgr = im_with_text[y, x]  # Note: OpenCV uses (y, x) for array indexing

        # Convert BGR to RGB format
        pixel_values_rgb = pixel_values_bgr[::-1]

        # Create a string with the pixel values
        pixel_values_str = f"RGB: {pixel_values_rgb}"

        # Draw text on the copied image
        cv2.putText(im_with_text, pixel_values_str, (x, y), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 255), 1)

        # Display the updated image
        cv2.imshow("Image", im_with_text)

# Create a window and set the mouse callback function
cv2.namedWindow("Image")
cv2.setMouseCallback("Image", on_mouse_event)

# Display the image
cv2.imshow("Image", im_cv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## IV. Is this a digital image? Explain why 

Yes, This is a digital image because it is a representation of a three dimensional image (scene) on a two-dimensional plane with three different channels Red, Green and Blue.

# 2. Convert the Picture from 1 above to gray color and display

```python
# Convert BGR image to grayscale
im_gray = cv2.cvtColor(im_cv, cv2.COLOR_BGR2GRAY)

# Save the grayscale image to a folder
cv2.imwrite("output_images/Arun1_grayscale.jpg", im_gray)

# Display the grayscale image
plt.imshow(im_gray, cmap='gray')
plt.axis('off')  # Hide axis
plt.show()
```

## I. Print the coordinates.

```python
height, width = im_gray.shape[:2]
channels = im_gray.shape[2] if len(im_gray.shape) == 3 else 1

# Print the width and height
print(f"Image Width: {width} pixels")
print(f"Image Height: {height} pixels")
print(f"Number of Channels: {channels}")
```

## II. Print the picture intensity

```python
# Get pixel intensity values in grayscale at the selected points
pixel1_intensity = im_gray[300, 300]
pixel2_intensity = im_gray[600, 400]

# Create a string with the pixel values
pixel1_values_str = f"Pixel intensity at 300, 300: {pixel1_intensity}"
print(pixel1_values_str)

pixel2_values_str = f"Pixel intensity at 400, 600: {pixel2_intensity}"
print(pixel2_values_str)
```

## III. Is this a digital image? Explain why

A grayscale image is also a digital image as it is represented on 2dimensional plane with only one plane which has a range of values from 0-255 that is shades of gray, where 0 is black and 255 is white.

## 3. Is there any difference between the co-ordinates from 1 and 2? Explain.

Yes, There is the difference between the co-ordinates from 1 & 2 (RGB-image and gray-scale image). 
Technically they might be same in width and height, But RGB image have 3 planes and Gray-scale image have only 1 plane

## 4. Is there any difference between the pixel values from 1 and 2? Explain.

Yes, There is a difference between pixel values of a Color image and Gray-scale image. 
Color image consists of three values as it's pixel value, they are R,G,B where each ranges from 0 to 255.
But Gray-scale image consists of only one value as its pixel value, which range from 0 to 255.
