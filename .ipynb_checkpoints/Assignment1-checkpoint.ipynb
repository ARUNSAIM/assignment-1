```python
import numpy as np
import cv2 
from PIL import Image, ImageFont, ImageDraw
from PIL.ImageChops import add, subtract, multiply, difference, screen
import PIL.ImageStat as stat

import matplotlib.pyplot as plt 
import matplotlib.image as img 
```

# 1. Take a head shot picture of yourself 

## I. Read and display the image 

```python
im_cv = cv2.imread("Arun1.jpg") #Read the image
#cv reads image as bgr by default
im = cv2.cvtColor(im_cv, cv2.COLOR_BGR2RGB)#convert from bgr to rgb and save
plt.imshow(im) 
```

## II. Print the coordinates.

```python
print(im.width, im.height, im.mode, im.format, type(im)) # Print Co-Ordinates and etc
plt.imshow(im)
```

## III. Print the picture intensity

```python
pixel_values = list(im.getdata()) # Store the pixel values into a list
print(len(pixel_values)) #length of the list
print(723*734) #length of list = X * Y
print(pixel_values[0:3]) # print the first 3 pixel 
print(pixel_values[-3:]) # print the last 3 pixel 
```

## Selecting a random pixels from image and show its values on image

```python
#Selecting a random pixels from image and show its values on image
im_draw=im.copy() #take the original image copy
draw = ImageDraw.Draw(im_draw) # Create a drawing object

# Take random point coordinates
random_x1 = 300
random_y1 = 300
pixel_value_at_x1_y1 = pixel_values[random_y1 * im.width + random_x1]
random_x2 = 400
random_y2 = 600
pixel_value_at_x2_y2 = pixel_values[random_y2 * im.width + random_x2]

point_size = 1  # We can adjust the point size as needed
draw.rectangle([random_x1 - point_size, random_y1 - point_size, random_x1 + point_size, random_y1 + point_size], fill='green')
draw.rectangle([random_x2 - point_size, random_y2 - point_size, random_x2 + point_size, random_y2 + point_size], fill='green')
#To make the point visible, I coloured its neighbours also

text_size = 25 
font = ImageFont.truetype("arial.ttf", text_size)  # You can change "arial.ttf" to the font file of your choice
text1 = f"Pixel Value: {pixel_value_at_x1_y1}"# get the values of the selected pixel from the list we stored
text2 = f"Pixel Value: {pixel_value_at_x2_y2}"# get the values of the selected pixel from the list we stored

draw.text((random_x1 + 10, random_y1), text1, font=font, fill='green') 
draw.text((random_x2 - 40, random_y2), text2, font=font, fill='green') 
# Adjust the position of text with the pixel values next to the point

# Display the modified image with selected pixel and its values on it
im_draw
```

```python
#To confirm the color of selected pixel, I am drawing 100*100 image with the selected pixel values
pixel_color1 = (180, 137, 121)
image1 = Image.new("RGB", (100, 100), pixel_color1)
# Display the image
image1
```

```python
pixel_color2 = (30, 21, 22)
image2 = Image.new("RGB", (100, 100), pixel_color2)
# Display the image
image2
```
